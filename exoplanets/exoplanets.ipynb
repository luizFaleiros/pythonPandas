```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import numpy as np
```

```python
planetas=sns.load_dataset('planets')
```

Veja as informações do dataset

```python
planetas.dropna(inplace=True)
```

Crie um pairplot dos dados quantitativos;

```python
sns.pairplot(planetas)
```

Crie um heatmap dos dados

```python
sns.heatmap(planetas.corr())
```

Divida os dados em conjunto de treinamento e teste;

```python
X = planetas[['number', 'mass', 'orbital_period',
               'year']]
y = planetas['distance']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

Crie o modelo de treinamento

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```
