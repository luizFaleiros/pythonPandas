```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# PONHA O SEU CÓDIGO AQUI,
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS
# ENCONTRAR.

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

# PONHA O SEU CÓDIGO AQUI,
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR

```python
sns.displot(titanic['fare'],bins=30,kde=False,color='red')
```

# PONHA O SEU CÓDIGO AQUI,
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

# PONHA O SEU CÓDIGO AQUI,
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.

```python
sns.swarmplot(x='class',y='age',s= 3,data=titanic,palette='Set1')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
